# House Sales Prediction In King County, USA

## The Data

We will be using data from a Kaggle data set:
https://www.kaggle.com/harlfoxem/housesalesprediction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
house_df = pd.read_csv('kc_house_data.csv')
```

## Exploratory Data Analysis

```python
house_df.info()
```

```python
house_df.isnull().sum()
```

```python
house_df.describe().transpose()
```

```python
plt.figure(figsize=(12,8))
sns.histplot(house_df['price'])
```

```python
sns.countplot(x=house_df['bedrooms'])
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='price', y='sqft_living', data=house_df)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='bedrooms',y='price',data=house_df)
```

## Geographical Properties

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='price',y='long',data=house_df)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='price',y='lat',data=house_df)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='long',y='lat',data=house_df,hue='price')
```
